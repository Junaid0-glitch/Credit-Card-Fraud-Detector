## Training the model with Imbalanced data

```python
X = df.drop("Class", axis=1)
y = df["Class"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=43, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
classifier = {
    "LogisticRegression" : LogisticRegression(),
    "RandomForestClassifier" : RandomForestClassifier(),
    "DecisionTreeClassifier" : DecisionTreeClassifier(),
}
```

```python
for name, model in classifier.items():
    print(f"\n========={name}=========")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_pred, y_test)
    print(report)
```

## Using Smote over sampling to get over with imbalanced data

```python
from imblearn.over_sampling import SMOTE
X_res, y_res = SMOTE().fit_resample(X, y)
```

```python
y_res.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, random_state=43, test_size=0.2)
```

```python
for name, model in classifier.items():
    print(f"\n========={name}===============")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_pred, y_test)
    print(report)
```

## Both Randomforest and decision tree are giving same results, but i would go with randomforest

> Random Forest is generally more robust, less prone to overfitting, and performs better on unseen data compared to a single Decision Tree.

```python
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)
report = classification_report(y_pred, y_test)
print(report)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)

plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve

probabilities = forest.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, probabilities)

fpr[:5], tpr[:5], thresholds[:5]
```

```python
from sklearn.metrics import auc

area = auc(fpr, tpr)
area
```

```python
gini = 2 * area - 1
print("Gini Coefficient:", gini)
```

```python
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % area)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
